```python
import numpy as n
from matplotlib import pyplot as plt
import pandas as p
```

```python
df =p.read_csv('covid19india.csv')
df
```

```python
data = df.iloc[:,1:]
data
```

```python
data.info()
```

# Inspect null values in each column

```python
data.isnull().sum().sort_values()/len(data)*100
```

# Total number of cases month wise

```python
# per day in month round(data.groupby('Month')['Num Cases'].sum()/30,2)
data.groupby('Month')['Num Cases'].sum()
```

```python
## no. of hospitalized cases per month
M=data[data['Current Status'] == "Hospitalized"].groupby('Month')['Num Cases'].sum()
M
```

```python
M.plot.bar()
plt.show
```

# Total Male/Female Infected 

```python
data.groupby('Gender')['Num Cases'].sum()
```

# Which age group is infected most?

```python
M=data.groupby('Age Bracket')['Num Cases'].sum().sort_values(ascending =False).head(10)
M
```

```python
M.plot.bar(figsize=(15,5))
plt.show
```

# State wise total number of cases

```python
M=data[data['Current Status'] == "Hospitalized"].groupby('Detected State')['Num Cases'].sum().sort_values(ascending = False)
```

```python
M.plot.bar(figsize=(15,5))
plt.show
```

# How many cases Everyday

```python
Day = data[data['Current Status'] == "Hospitalized"].groupby(['Month','Date'])[['Num Cases']].sum()
Day
```

```python
Day.unstack(level =0).plot(kind ='bar',subplots = True, figsize =(15,14))
plt.show
```
